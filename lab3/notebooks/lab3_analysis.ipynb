```python
from data.data_processing import prepare_data

X, y, feature_columns = prepare_data()
print(X.head())
print(y.head())
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)
print(X_train.shape, X_test.shape)
```

```python
from models.knn_model import make_knn_pipeline
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

baseline_model = make_knn_pipeline(n_neighbors=5, weights="distance", p=2)
baseline_model.fit(X_train, y_train)
predictions = baseline_model.predict(X_test)
print({
    "MAE": mean_absolute_error(y_test, predictions),
    "RMSE": mean_squared_error(y_test, predictions) ** 0.5,
    "R2": r2_score(y_test, predictions),
})
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, KFold, TimeSeriesSplit
import numpy as np

param_grid = {
    "knn__n_neighbors": list(range(2, 26)),
    "knn__weights": ["uniform", "distance"],
    "knn__p": [1, 2],
}
random_distributions = {
    "knn__n_neighbors": np.arange(2, 26),
    "knn__weights": ["uniform", "distance"],
    "knn__p": [1, 2],
}
cv_strategies = {
    "KFold": KFold(n_splits=5, shuffle=False),
    "TimeSeriesSplit": TimeSeriesSplit(n_splits=5),
}
for cv_name, cv in cv_strategies.items():
    search = GridSearchCV(make_knn_pipeline(), param_grid=param_grid, cv=cv, scoring="neg_mean_absolute_error", n_jobs=1)
    search.fit(X_train, y_train)
    print(cv_name, search.best_params_, -search.best_score_)
```

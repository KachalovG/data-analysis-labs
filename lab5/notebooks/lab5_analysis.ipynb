## Описание задачи

Необходимо спрогнозировать общий индекс потребительских цен `total` с помощью ансамблевых методов. Для сравнения выбраны две модели группы бэггинга, а также `AdaBoost` и `Gradient Boosting`.


```python
from data.data_processing import prepare_data

X, y, feature_columns = prepare_data()
print(X.head())
print(y.head())
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)
print(X_train.shape, X_test.shape)
```

```python
from models.ensemble_models import get_models
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

rows = []
for name, model in get_models().items():
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    rows.append({
        'Model': name,
        'MAE': mean_absolute_error(y_test, pred),
        'RMSE': mean_squared_error(y_test, pred) ** 0.5,
        'R2': r2_score(y_test, pred),
    })
print(rows)
```

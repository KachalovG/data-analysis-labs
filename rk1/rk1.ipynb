## Исходные данные

Использован набор данных `iris` из `scikit-learn`. В документации `load_iris` указано, что это классический набор данных для многоклассовой классификации: 3 класса, 150 объектов и 4 количественных признака.


## Постановка задачи

Так как в исходном `iris` нет категориальных признаков и пропусков, они создаются искусственно. Добавляется категориальный признак `petal_length_category`, а также вносятся пропуски в один количественный и один категориальный признак.


```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris(as_frame=True)
df = iris.frame.copy()
df['species_name'] = df['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
df.head()
```

```python
df['petal_length_category'] = pd.cut(
    df['petal length (cm)'],
    bins=[0, 2.5, 5.0, 10.0],
    labels=['short', 'medium', 'long'],
    include_lowest=True
).astype('object')

df.loc[[3, 14, 27, 48, 59, 76, 89, 101, 118, 134], 'sepal width (cm)'] = float('nan')
df.loc[[5, 16, 33, 57, 72, 94, 111, 145], 'petal_length_category'] = float('nan')

df[['sepal width (cm)', 'petal_length_category']].isna().sum()
```

## Обработка пропусков

Для количественного признака `sepal width (cm)` использована замена медианой (`median`). Для категориального признака `petal_length_category` использована замена наиболее частым значением (`most_frequent`).


```python
from sklearn.impute import SimpleImputer

processed_df = df.copy()
processed_df[['sepal width (cm)']] = SimpleImputer(strategy='median').fit_transform(processed_df[['sepal width (cm)']])
processed_df[['petal_length_category']] = SimpleImputer(strategy='most_frequent').fit_transform(processed_df[['petal_length_category']])
processed_df[['sepal width (cm)', 'petal_length_category']].isna().sum()
```

## ?????????? ??????

???????????? ????? ?????? `iris` ?? `scikit-learn`. ??? ??? ? ???????? ?????? ??? ?????????????? ????????? ? ?????????, ? ?????? ???????????? ????????? ?????????????? ??????? `petal_length_group`, ? ????? ??????????? ???????? ? ???? ?????????????? ? ???? ?????????????? ???????.

```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris(as_frame=True)
df = iris.frame.copy()
df["species"] = df["target"].map({0: "setosa", 1: "versicolor", 2: "virginica"})
df.head()
```

```python
df["petal_length_group"] = pd.cut(
    df["petal length (cm)"],
    bins=[0, 2.5, 5.0, 10.0],
    labels=["short", "medium", "long"],
    include_lowest=True
).astype("object")

df.loc[[3, 14, 27, 48, 59, 76, 89, 101, 118, 134], "sepal width (cm)"] = float("nan")
df.loc[[5, 16, 33, 57, 72, 94, 111, 145], "petal_length_group"] = float("nan")

df[["sepal width (cm)", "petal_length_group"]].isna().sum()
```

## ????????? ?????????

??? ??????????????? ???????? `sepal width (cm)` ???????????? ?????? ???????? (`median`). ??? ??????????????? ???????? `petal_length_group` ???????????? ?????? ???????? ?????? ????????? (`most_frequent`).

```python
from sklearn.impute import SimpleImputer

processed = df.copy()
processed[["sepal width (cm)"]] = SimpleImputer(strategy="median").fit_transform(processed[["sepal width (cm)"]])
processed[["petal_length_group"]] = SimpleImputer(strategy="most_frequent").fit_transform(processed[["petal_length_group"]])
processed[["sepal width (cm)", "petal_length_group"]].isna().sum()
```
